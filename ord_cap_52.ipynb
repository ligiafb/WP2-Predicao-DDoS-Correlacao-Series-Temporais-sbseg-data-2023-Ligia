# Loop para extrair as métricas sem variar o taux

```python
# Matriz com as métricas
matrix = []
label = []

# Definição do Tau e Dx
taux = 1
dx = 3

# Time set
init = 0
end = 48#440 #(5%)

for i in range (1, end):
    matrix.append([[init, i]])

for i in range(0,925): #972-48+1
    # Vetor para guardar as métricas
    vec = []
    vec.append([init, end])
    # COLUNA TOTAL_IPS_ORIGEM

    # Calculo da Entropia e da Complexidade
    tmp = ordpy.complexity_entropy(df['total_ips_origem'][init:end], dx=dx, taux=taux)
    vec.append(tmp[0])
    vec.append(tmp[1])

     #Calculo da Informação de Fisher Shannon
    tmp = ordpy.fisher_shannon(df['total_ips_origem'][init:end], dx=dx, taux=taux)
    vec.append(tmp[1])

    # Transformação em grafo de transição
    edges = ordpy.ordinal_network(df['total_ips_origem'][init:end], dx=dx, taux=taux)

    # Entropia e complexidade do grafo
    tmp = ordpy.complexity_entropy(edges[2], dx=dx, taux=taux)
    vec.append(tmp[0])
    vec.append(tmp[1])

    # Informação de fisher do grafo
    tmp = ordpy.fisher_shannon(edges[2], dx=dx, taux=taux)
    vec.append(tmp[1])

    # Quantidade de arestas
    tmp = len(edges[1])
    vec.append(tmp)

    # Extração dos índices das autotransições
    j=0
    indices = []
    for array in edges[1]:
        if(array[0] == array[1]):
            indices.append(j)
        j += 1

    # Calculo da probabilidade de autotransição
    prob = 0
    for index in indices:
        tmp = edges[2][index]
        prob += tmp
    vec.append(prob)

    # COLUNA TOTAL_IPS_DESTINO

    # Calculo da Entropia e da Complexidade
    tmp = ordpy.complexity_entropy(df['total_ips_destino'][init:end], dx=dx, taux=taux)
    vec.append(tmp[0])
    vec.append(tmp[1])

    #Calculo da Informação de Fisher Shannon
    tmp = ordpy.fisher_shannon(df['total_ips_destino'][init:end], dx=dx, taux=taux)
    vec.append(tmp[1])

    # Transformação em grafo de transição
    edges = ordpy.ordinal_network(df['total_ips_destino'][init:end], dx=dx, taux=taux)

    # Entropia e complexidade do grafo
    tmp = ordpy.complexity_entropy(edges[2], dx=dx, taux=taux)
    vec.append(tmp[0])
    vec.append(tmp[1])

    # Informação de fisher do grafo
    tmp = ordpy.fisher_shannon(edges[2], dx=dx, taux=taux)
    vec.append(tmp[1])

    # Quantidade de arestas
    tmp = len(edges[1])
    vec.append(tmp)

    # Extração dos índices das autotransições
    j=0
    indices = []
    for array in edges[1]:
        if(array[0] == array[1]):
            indices.append(j)
        j += 1

    # Calculo da probabilidade de autotransição
    prob = 0
    for index in indices:
        tmp = edges[2][index]
        prob += tmp
    vec.append(prob)

    # COLUNA TOTAL_PACOTES

    # Calculo da Entropia e da Complexidade
    tmp = ordpy.complexity_entropy(df['total_pacotes'][init:end], dx=dx, taux=taux)
    vec.append(tmp[0])
    vec.append(tmp[1])

    #Calculo da Informação de Fisher Shannon
    tmp = ordpy.fisher_shannon(df['total_pacotes'][init:end], dx=dx, taux=taux)
    vec.append(tmp[1])

    # Transformação em grafo de transição
    edges = ordpy.ordinal_network(df['total_pacotes'][init:end], dx=dx, taux=taux)

    # Entropia e complexidade do grafo
    tmp = ordpy.complexity_entropy(edges[2], dx=dx, taux=taux)
    vec.append(tmp[0])
    vec.append(tmp[1])

    # Informação de fisher do grafo
    tmp = ordpy.fisher_shannon(edges[2], dx=dx, taux=taux)
    vec.append(tmp[1])

    # Quantidade de arestas
    tmp = len(edges[1])
    vec.append(tmp)

    # Extração dos índices das autotransições
    j=0
    indices = []
    for array in edges[1]:
        if(array[0] == array[1]):
            indices.append(j)
        j += 1

    # Calculo da probabilidade de autotransição
    prob = 0
    for index in indices:
        tmp = edges[2][index]
        prob += tmp
    vec.append(prob)


    # Adiciona as métricas a matriz
    matrix.append(vec)

    # Vetor label
    bot = 0
    for line in df.iloc[init:end].iterrows():
        if(line[1][4] == 1):
            bot = 1
    label.append(bot)
        
     # Loop de tempo
    init = init + 1
    end = end + 1

    
# Cria um dataframe com a matriz
ft = pd.DataFrame(matrix, columns = ['time',
                                     'IPO_entropy', 'IPO_complexity' , 'IPO_fisher_shanon', 'IPO_edge_entropy', 'IPO_edge_complexity', 'IPO_edge_fisher_shannon', 'IPO_number_nodes', 'IPO_self_transition',
                                     'IPD_entropy', 'IPD_complexity' , 'IPD_fisher_shanon', 'IPD_edge_entropy', 'IPD_edge_complexity', 'IPD_edge_fisher_shannon', 'IPD_number_nodes', 'IPD_self_transition',
                                     'TP_entropy', 'TP_complexity' , 'TP_fisher_shanon', 'TP_edge_entropy', 'TP_edge_complexity', 'TP_edge_fisher_shannon', 'TP_number_nodes', 'TP_self_transition'])
ft
ft.to_csv('po_52.csv',sep=';')
```

```python
from sklearn.svm import OneClassSVM
from sklearn.metrics import confusion_matrix
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score

po = pd.read_csv("po_52.csv", sep=";")
early_warning = pd.read_csv("cap_52.csv", sep=";")

colunas = ['total_ips_origem', 'total_ips_destino', 'total_pacotes']

df = po[['IPO_entropy', 'IPO_complexity',
       'IPO_fisher_shanon', 'IPO_edge_entropy', 'IPO_edge_complexity',
       'IPO_edge_fisher_shannon', 'IPO_number_nodes', 'IPO_self_transition',
       'IPD_entropy', 'IPD_complexity', 'IPD_fisher_shanon',
       'IPD_edge_entropy', 'IPD_edge_complexity', 'IPD_edge_fisher_shannon',
       'IPD_number_nodes', 'IPD_self_transition', 'TP_entropy',
       'TP_complexity', 'TP_fisher_shanon', 'TP_edge_entropy',
       'TP_edge_complexity', 'TP_edge_fisher_shannon', 'TP_number_nodes',
       'TP_self_transition']]
label = early_warning['has_bot']

init = 48
train = 324
end = 778

x_train = df[init:train]
x_test = df[train:end]
y_real = label[train:end]

clf = OneClassSVM(gamma='auto', kernel='poly', nu=0.1).fit(x_train)
y_test = clf.predict(x_test)
y_test_final = [0 if i == 1 else 1 for i in y_test]
acc = accuracy_score(y_real, y_test_final)
print(acc)
print(confusion_matrix(y_real, y_test_final))
print(classification_report(y_real, y_test_final))
```
